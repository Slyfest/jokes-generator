```python
import pandas as pd
import markovify
```

```python
jokes_df = pd.read_csv("data/jokes.csv")
```

```python
jokes_df.groupby(["category"]).count()
```

```python
momma_jokes = jokes_df[jokes_df["category"] == "Yo momma humor"]
text_model = markovify.NewlineText(momma_jokes["text"], state_size=2)
```

```python
for i in range(20):
    print(text_model.make_sentence())
```
